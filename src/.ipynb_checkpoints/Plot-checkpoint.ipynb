```python
t = np.loadtxt('../examples/1d/timings.txt')
print(t.shape)

fig, ax = plt.subplots(figsize = (10, 6))
#ax.plot(x, n)
#ax.plot(x, ux)
ax.plot(t[:, 0], label="RECONSTRUCTION")
ax.plot(t[:, 1], label="BOUNDARY_CONDITIONS")
ax.plot(t[:, 2], label="FLUXES")
ax.plot(t[:, 3], label="RHS")
ax.plot(t[:, 4], label="UPDATE")
plt.legend()
```
